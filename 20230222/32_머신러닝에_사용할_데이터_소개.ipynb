```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets # 사이킷런 라이브러리에서 제공하는 데이터셋
```

머신러닝에 사용할 데이터 소개

집값 예측하기

```python
# 사이킷런 라이브러리에서 보스턴 집값 데이터를 불러온다.
# 사이킷런에서 제공하는 데이터는 data와 target으로 구성되어 있다.
raw_boston = datasets.load_boston()
# print(raw_boston)
X_boston = pd.DataFrame(raw_boston.data) # 피쳐 부분을 추출한다.
y_boston = pd.DataFrame(raw_boston.target) # 타겟 부분을 추출한다.
# 피쳐와 타겟 데이터를 결합해 전체 데이터프레임을 생성한다.
# pandas 라이브러리의 concat() 함수의 axis 속성은 결합 방향을 의미하는데 생략하거나 axis=0으로 설정하면 주어진
# 데이터를 위 + 아래 형태로 합치고 axis=1로 설정하면 좌 + 우 형태로 합친다.
df_boston = pd.concat([X_boston, y_boston], axis=1)
df_boston
```

```python
feature_boston = raw_boston.feature_names
print(feature_boston)
```

```python
col_boston = np.append(feature_boston, ['target'])
print(col_boston)
# 피쳐와 타겟 데이터를 결합해 생성한 데이터프레임에 열 이름을 적용한다.
df_boston.columns = col_boston
df_boston.head()
```
