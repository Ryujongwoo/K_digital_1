```python
import warnings
warnings.filterwarnings(action='ignore')
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By # selenium 4 버전 'By'를 사용하기 위해 import 한다.
import time
```

from selenium.webdriver.common.by import By를 import 하고 사용한다.  
driver.find_element_by_class_name('') => driver.find_element(By.CLASS_NAME, '')로 변경됨  
driver.find_element_by_id('') => driver.find_element(By.ID, '')로 변경됨  
driver.find_element_by_name('') => driver.find_element(By.NAME, '')로 변경됨  
driver.find_element_by_link_text('') => driver.find_element(By.LINK_TEXT, '')로 변경됨  
driver.find_element_by_partial_link_text('') => driver.find_element(By.PARTIAL_LINK_TEXT, '')로 변경됨  
driver.find_element_by_tag_name('') => driver.find_element(By.TAG_NAME, '')로 변경됨  
driver.find_element_by_css_selector('') => driver.find_element(By.CSS_SELECTOR, '')로 변경됨  
driver.find_element_by_xpath('') => driver.find_element(By.XPATH, '')로 변경됨

```python
option = Options()
option.add_argument('--window-size=1890,1030')
driver = webdriver.Chrome('./chromedriver.exe', options=option)
driver.implicitly_wait(5)

driver.get('https://comic.naver.com/comment/comment?titleId=790713&no=89')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
reviews = soup.select('span.u_cbox_contents')
for review in reviews[1:2]:
    print(review.text)
    
# 답글 버튼을 클릭한다.
# driver.find_element_by_xpath('element_xpath')가 driver.find_element(By.XPATH, 'element_xpath')로 변경됨
# driver.find_element_by_xpath( # V3
driver.find_element(By.XPATH, 
    # '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[1]/div[1]/div/div[5]/a' # 1번째 베댓의 댓글 버튼
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[1]/div/div[5]/a' # 2번째 베댓의 댓글 버튼
    # '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[3]/div[1]/div/div[5]/a' # 3번째 베댓의 댓글 버튼
    # '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[4]/div[1]/div/div[5]/a' # 4번째 베댓의 댓글 버튼
    # ...
    # '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[15]/div[1]/div/div[5]/a' # 15번째 베댓의 댓글 버튼
).click()
time.sleep(1)
print('베댓', '=' * 80)

# selenium 새로고침 방법
# driver.refresh() # 새로고침을 실행한다.
# driver.get(driver.current_url) # get() 함수에 현재 페이지 주소를 넣어주는 방법으로 새로고침을 할 수 있다.
# execute_script() 함수로 자바스크립트 코드를 실행시킬 수 있다.
# driver.execute_script('location.reload()') # 자바스크립트 location 객체의 reload() 함수를 실행해서 새로고침을 할 수 있다.

# 페이지 내용이 변경되면 다시 읽어와야 한다.
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('베댓 댓글 1페이지', '=' * 80)

driver.find_element(By.XPATH, 
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[1]'
).click()
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('베댓 댓글 2페이지', '=' * 80)

driver.find_element(By.XPATH, 
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[3]'
).click()
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('베댓 댓글 3페이지', '=' * 80)

driver.find_element(By.XPATH, 
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[4]'
).click()
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('베댓 댓글 4페이지', '=' * 80)

driver.find_element(By.XPATH, 
    '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[2]/div[2]/div[1]/div/a[5]'
    
).click()
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
print(len(replys))
for reply in replys:
    print(reply.text)
print('베댓 댓글 5페이지', '=' * 80)
```
