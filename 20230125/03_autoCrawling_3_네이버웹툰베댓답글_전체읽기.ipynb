```python
import warnings
warnings.filterwarnings(action='ignore')
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
```

```python
option = Options()
option.add_argument('--window-size=1890,1030')
driver = webdriver.Chrome('./chromedriver.exe', options=option)
driver.implicitly_wait(5)

driver.get('https://comic.naver.com/comment/comment?titleId=650305&no=369')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
reviews = soup.select('span.u_cbox_contents')

# 베댓의 개수만큼 반복한다.
for i in range(len(reviews[:])):
    print('{:02d}번째 베댓\n{}'.format(i + 1, reviews[i].text))
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    # 베댓 답글을 연다.
    path = '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[' + str(i + 1) + ']/div[1]/div/div[4]/a'
    driver.find_element_by_xpath(path).click()
    time.sleep(1)
    
    # 베댓 1번째 답글 페이지를 읽는다.
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
    print('{:d}번째 베댓 페이지 답글 개수: {}개'.format(i + 1, len(replys)))
    
    # 베댓 1번째 답글 페이지에 답글이 있으면 출력한다.
    if len(replys) == 0:
        print('답글 없음')
    else:
        for reply in replys:
            print(reply.text)
    print()
    
    # 베댓 1번째 답글 페이지를 제외한 나머지 답글 페이지 개수를 얻어온다.
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    page = soup.select('.u_cbox_reply_area .u_cbox_page')
    print('베댓 댓글 페이지 개수: {}\n'.format(len(page)))
    
    # 1번째 베댓 페이지의 댓글은 읽어왔으므로 2번째 베댓 댓글 페이지 부터 마지막 베댓 댓글 페이지까지 반복한다.
    for j in range(1, len(page)):
        # 다음에 읽을 베댓 답글을 연다.
        driver.find_element_by_css_selector(
            '#cbox_module_wai_u_cbox_content_wrap_tabpanel div.u_cbox_paginate > div > a:nth-child(' + str(j + 3) + ')'
        ).click()
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        replys = soup.select('.u_cbox_reply_area .u_cbox_contents')
        for reply in replys:
            print(reply.text)
    
    # 베댓 답글을 닫는다.
    driver.find_element_by_xpath(path).click()
    time.sleep(1)
```
