```python
# !pip install requests # 크롤링할 사이트에 접속해서 웹 문서를 읽어온다.
# !pip install beautifulsoup4 # requests를 사용해서 읽어온 웹 문서를 파싱(분석)한다.
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'http://dowellcomputer.com/main.jsp'
```

```python
# requests 모듈의 get() 메소드로 타겟 사이트의 정보를 얻어온다. 요청한다.
req = requests.get(targetSite)
print(req)
# <Response [200]>: 서버가 요청을 제대로 처리했다는 뜻으로 서버가 요청한 페이지를 제공했다는 의미로 쓰인다.
```

```python
# 타겟 사이트에서 얻어온 정보 중에서 text로 된 html 문서만 얻어온다.
html = req.text
print(html)
```

```python
# bs4 모듈의 BeautifulSoup() 함수로 타겟 사이트에서 얻어온 html 문서를 html parser를 사용해서 파싱할 수 있도록 준비한다.
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

# html 태그와 html 태그에 지정한 속성을 이용한 크롤링
findAll('태그이름', {'속성이름': '속성값', ...}) => 속성없이 태그 단위로 크롤링할 때는 {} 안의 내용을 생략할 수 있다.

```python
images = soup.findAll('img')
print(images)
print(type(images))
for image in images:
    print(image)
```

```python
# 최근 공지사항의 제목은 <b> 태그 내부에 있다.
notices = soup.findAll('b')
# print(notices)
for notice in notices:
    print(notice.text)
```

```python
# 최근 공지사항의 제목은 class 속성값이 'tail'인 <td> 태그 내부에 있다.
notices = soup.findAll('td', {'class': 'tail'})
# print(notices)
for notice in notices:
    print(notice)
```

```python
# 최근 공지사항의 제목은 href 속성의 속성값에 'notice' 문자열이 포함되어있는 <a> 태그 내부에 있다.
notices = soup.select('td > a')
# print(notices)
for notice in notices:
    # print(notice)
    # get('속성이름') 메소드로 인수로 지정한 속성의 속성값을 얻어온다.
    print(notice.get('href'))
    # 'href' 속성의 속성값 중에서 'notice'가 포함된 요소만 얻어온다.
    if notice.get('href').find('notice') >= 0:
        print(notice.text)
```
