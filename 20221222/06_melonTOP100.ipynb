# Trident User Agents
https://developers.whatismybrowser.com/useragents/explore/layout_engine_name/trident/

```python
targetSite = 'https://www.melon.com/chart/index.htm'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get(targetSite, headers=header)
print(req)
```

```python
html = req.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
# print(titles)
# print(len(titles))

for i in range(len(titles)):
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.strip()))
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[1]))
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0]))
```

```python
# 아트스트 크롤링
artists = soup.findAll('span', {'class': 'checkEllipsis'})
# print(artists)
# print(len(artists))

for i in range(len(artists)):
    # print(artists[i].text.split('\n'))
    # print('{:3d}위 {}'.format(i + 1, artists[i].text))
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.strip()))
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[0]))
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
    print('{:3d}위 {}'.format(i + 1, artists[i].text.split(',')[0]))
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.split(',')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```

```python
with open('./output/melonTOP100.txt', 'w') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(100):
        artist = artists[i].text.split(',')[0]
        title = titles[i].text.strip().split('\n')[0]
        file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
print('melonTOP100.txt 파일로 쓰기 완료')
```
