concat() 함수는 연결하려는 데이터프레임을 리스트에 담아 인수로 전달하면 연결된 데이터프레임을 반환한다.  
concat([데이터프레임, 데이터프레임, ...])  
데이터프레임을 연결할 때 기본값이 axis=0으로 설정되어있어 위에서 아래 방향으로 연결된다.

```python
# df1, df2, df3은 열 이름이 모두 A, B, C, D로 같고 연결한 다음에도 열이 그대로 유지되고 인덱스도 그대로 유지된다.
# row_concat = pd.concat([df1, df2, df3])
row_concat = pd.concat([df1, df2, df3], axis=0)
row_concat
```

```python
# 데이터프레임에 시리즈 연결하기
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
new_row_series
```

```python
# 시리즈가 새로운 행으로 추가되지 않고 새로운 열로 추가된다. => NaN 이라는 값이 많이 생긴다. => NaN을 누락값(결측치)이라고
# 부른다.
# 시리즈에 열 이름이 없기 때문에 시리즈를 데이터프레임의 새로운 행으로 추가하려할 때 제대로 추가되지 않는다.
# 새로운 시리즈를 새로운 열로 간주해서 0이라는 이름의 열로 추가된다.
pd.concat([df1, new_row_series])
```

concat() 함수는 하나의 데이터프레임에 한 번에 2개 이상이 데이터프레임을 연결하는 함수이다. 만약에 데이터프레임에 연결할 데이터프레임이 1개라면 append() 메소드를 사용해도 된다.  
데이터프레임.append(추가할 데이터프레임)

```python
df1.append(new_row_df)
```

```python
# concat() 함수나 append() 함수를 사용해서 데이터프레임을 연결할 때 인덱스를 다시 부여하고 싶다면 ignore_index=True 옵션을
# 지정하면 된다.
pd.concat([df1, new_row_df], ignore_index=True)
```

```python
df1.append(new_row_df, ignore_index=True)
```

append() 함수와 딕셔너리를 사용하면 추가할 데이터를 데이터프레임으로 구성하지 않고도 데이터프레임에 추가할 수 있다.  
데이터프레임.append(추가할 데이터가 저장된 딕셔너리)

```python
data_dict = {'A': 'n5', 'B': 'n6', 'C': 'n7', 'D': 'n8'}
print(data_dict)
# 데이터프레임이 아닌 딕셔너리를 추가하기 때문에 인덱스가 없으므로 ignore_index=True 옵션을 지정해서 기존 인덱스를 무시하고
# 다시 부어되게 해야한다.
# df1.append(data_dict) # 에러
df1.append(data_dict, ignore_index=True)
```

```python
# ignore_index=True 옵션의 기능은 데이터프레임을 연결하고 인덱스를 0부터 다시 부여한다.
row_concat = pd.concat([df1, df2, df3], ignore_index=True)
row_concat
```

concat() 함수의 join 속성을 inner로 지정하면 데이터프레임에 공통으로 데이터가 있는 열만 골라서 연결할 수 있고 누락값이 발생되지 않는다.

```python
# join 속성의 기본값은 outer이고 공통으로 데이터가 있는 열에 신경쓰지 않고 모두 연결한다. => NaN이 대량 발생된다.
col_concat = pd.concat([df1, df2, df3], ignore_index=True, join='outer')
col_concat
```

```python
# df1, df2, df3 데이터프레임은 3개 모두 공통 열이 없기 때문에 빈 데이터프레임이 생성된다.
col_concat = pd.concat([df1, df2, df3], ignore_index=True, join='inner')
col_concat
```

```python
# df1, df3 데이터프레임은 2개 모두 공통 열이 있기 때문에 공통 열을 연결한 데이터프레임이 생성된다. => A, C열이 연결된다.
col_concat = pd.concat([df1, df3], ignore_index=True, join='inner')
col_concat
```

```python
# df2, df3 데이터프레임은 2개 모두 공통 열이 있기 때문에 공통 열을 연결한 데이터프레임이 생성된다. => F, H열이 연결된다.
col_concat = pd.concat([df2, df3], ignore_index=True, join='inner')
col_concat
```

merge() 함수는 기본적으로 내부 조인을 실행하며 함수를 실행한 데이터프레임이 왼쪽으로 지정되고 첫 번째 인수로 지정된 데이터프레임이 오른쪽으로 지정되며 left_on 속성과 right_on 속성은 각각 왼쪽과 오른쪽 데이터프레임에서 일치해야 할 열을 지정한다.

```python
# survey_person.csv => 관측한 사람
# survey_site.csv => 관측 위치
# survey_survey.csv => 기상 정보
# survey_visited.csv => 관측 날짜
```

```python
person = pd.read_csv('./data/survey_person.csv')
person
```

```python
site = pd.read_csv('./data/survey_site.csv')
site
```

```python
survey = pd.read_csv('./data/survey_survey.csv')
survey
```

```python
visited = pd.read_csv('./data/survey_visited.csv')
visited
```

```python
print(site.columns)
print(visited.columns)
# how 속성으로 두 데이터프레임을 합치는 방법을 지정한다. 기본값은 how='inner'이다.
# how 속성을 생략하거나 how='inner'(내부 조인)로 지정하면 두 데이터프레임에 공통적으로 포함된 값이 있는 데이터만 합친다.
# how 속성을 how='outer'(외부 조인)로 지정하면 두 데이터프레임의 모든 데이터를 합친다. 한쪽에만 있는 데이터는 NaN이 채워진다.
# how 속성을 how='left'(왼쪽 외부 조인)로 지정하면 왼쪽 데이터프레임의 모든 데이터와 오른쪽 데이터프레임에 있는 공통적으로
# 포함된 데이터를 합친다. 한쪽에만 있는 데이터는 NaN이 채워진다.
# how 속성을 how='right'(오른쪽 외부 조인)로 지정하면 오른쪽 데이터프레임의 모든 데이터와 왼쪽 데이터프레임에 있는 공통적으로
# 포함된 데이터를 합친다. 한쪽에만 있는 데이터는 NaN이 채워진다.
s2v_merge = site.merge(visited, left_on='name', right_on='site', how='right')
s2v_merge
```

```python
print(person.columns)
print(survey.columns)
p2s_merge = person.merge(survey, left_on='ident', right_on='person', how='left')
p2s_merge
```

```python
print(visited.columns)
print(survey.columns)
v2s_merge = visited.merge(survey, left_on='ident', right_on='taken', how='inner')
v2s_merge
```

```python
print(p2s_merge.columns)
print(v2s_merge.columns)
# left_on, right_on 속성에 전달하는 열 데이터는 여러개를 사용할 수 있다. => 여러개의 열 이름을 리스트에 담아서 전달한다.
p2s_v2s_merge = p2s_merge.merge(v2s_merge, left_on=['ident', 'taken', 'quant', 'reading'], 
                                right_on=['person', 'ident', 'quant', 'reading'])
# merge() 함수를 실행하는 두 데이터프레임에 공통된 열 이름이 있으면 _x, _y 접미사가 추가된다.
# _x는 왼쪽 데이터프레임을 의미하고 _y는 오른쪽 데이터프레임을 의미한다.
p2s_v2s_merge
```
