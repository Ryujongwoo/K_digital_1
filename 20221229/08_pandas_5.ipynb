```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
```

```python
# Pew Research Center는 세계의 여론 조사, 인구 통계 등 다양한 사회 과학 연구를 수행하는 곳이다.
# Pew Research Center에서 조사한 '미국의 소득과 종교' 데이터를 사용한다.
pew = pd.read_csv('./data/pew.csv')
pew
```

melt() 함수  
데이터프레임을 깔끔한 데이터로 정리하는데 사용하는 melt() 함수는 지정한 열의 데이터를 모두 행으로 정리해준다.  
melt(데이터프레임, id_vars[, value_vars, var_name, value_name])  
id_vars => 위치를 그대로 유지할 열 이름을 지정한다. 필수  
value_vars => 행으로 위치를 변경할 열 이름을 지정한다.  
var_name => value_vars로 위치를 변경한 열 이름을 지정한다.  
value_name => value_vars로 위치를 변경한 열의 데이터를 저장한 열의 이름을 지정한다.

```python
# 소득 정보가 열을 구성하고 있다. => 소득 정보 열을 행 데이터로 옮기려고 한다.
# id_vars 속성으로 지정한 열(religion)을 제외한 나머지 소득 정보 열(<$10k, $10-20k, ..., >150k, Don't know/refused)이
# variable 열로 소득 정보 열의 데이터는 value로 정리된다.
# value_vars 속성을 생략하면 모든 열을 대상으로 함수가 실행된다.
# 이러한 과정을 'religion' 열을 고정하여 피벗했다. 라고 말한다.
pew_long = pd.melt(pew, id_vars='religion')
pew_long.head(40)
```

```python
# value_vars 속성을 지정하면 지정한 열만 대상으로 피벗한다.
pew_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k')
pew_long
```

```python
# 피벗할 열이 2개 이상일 경우 피벗할 열을 리스트로 묶어서 value_vars 속성으로 전달하면 된다.
pew_long = pd.melt(pew, id_vars='religion', value_vars=['$10-20k', '$100-150k'])
pew_long
```

```python
# var_name 속성을 이용해서 variable로 표시되던 열 이름을 '연봉'으로 변경하고 value_name 속성을 이용해서 value로 표시되던
# 열 이름을 '인원수'로 변경한다.
pew_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k', var_name='연봉', value_name='인원수')
pew_long
```

```python
pew_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k')
pew_long.columns = ['종교', '연봉', '인원수']
pew_long
```

```python
ebola = pd.read_csv('./data/country_timeseries.csv')
ebola
```

```python
ebola_short = ebola.loc[:, ['Date', 'Day', 'Cases_Guinea', 'Cases_Liberia', 'Deaths_Guinea', 'Deaths_Liberia']]
ebola_short
```

```python
ebola_short = ebola.iloc[:, [0, 1, 2, 3, 10, 11]]
ebola_short
```

```python
# Date와 Day 열을 고정하고 나머지를 행으로 피벗하면 각 나라별 발병자 수와 사망자 수를 볼 수 있다.
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long
```

```python
# split() 함수에 '_'를 구분자로 전달하면 Cases_Guinea를 Cases와 Guinea로 분리할 수 있다.
print(type(ebola_long.variable)) # <class 'pandas.core.series.Series'>
# 시리즈 뒤에 '.'을 찍고 str을 붙여주면 시리즈에 문자열 관련 함수를 실행할 수 있도록 해준다.
print(type(ebola_long.variable.str)) # <class 'pandas.core.strings.StringMethods'>
variable_split = ebola_long.variable.str.split('_')
print(type(variable_split)) # class 'pandas.core.series.Series'> 분리된 데이터 전체는 시리즈 타입의 데이터이다.
print(variable_split)
# 분리된 데이터가 저장된 시리즈에서 특정 위치의 데이터만 얻어오면 리스트 타입의 데이터이다.
print(type(variable_split[0])) # <class 'list'>
print(variable_split[0])
print(variable_split[0][1])
```

```python
# status_values = variable_split[0] # variable_split 시리즈의 0번째 인덱스 위치의 값 [Cases, Guinea]를 얻어온다.
# status_values = variable_split[0][0] # 0번째 인덱스 위치의 값 [Cases, Guinea]에서 0번째 인덱스인 'Cases'를 얻어온다.
# str을 적용해서 문자열 관련 기능이 실행되면 다시 시리즈로 돌아가기 때문에 문자열 관련 기능을 다시 실행하려면 str을 다시
# 적용해야 한다.
status_values = variable_split.str[0]
print(type(status_values))
print(status_values)
```

```python
status_values = variable_split.str.get(0)
print(type(status_values))
print(status_values)
```

```python
country_values = variable_split.str[1]
print(country_values)
```

```python
print(ebola_long.columns)
ebola_long['status'] = status_values
ebola_long['country'] = country_values
ebola_long
```

```python
# split() 함수에 expand='True' 옵션을 지정하면 split() 함수가 실행된 결과를 데이터프레임으로 만든다.
variable_split = ebola_long.variable.str.split('_', expand=True)
variable_split
```

```python
variable_split.columns = ['status2', 'country2']
variable_split
```

```python
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)
ebola_parsed
```
